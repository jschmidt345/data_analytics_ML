```python
def pearson(arr1, arr2, return_scatter=False):
    """
    Calculates pearson coefficient (R^2)
    Paramaters
    ----------
    arg1 : type array
    arg2 : type array
    """
    arr1 = np.array(arr1)
    arr2 = np.array(arr2)
    
    numerator = (len(arr1) * np.sum(arr1 * arr2) - np.sum(arr1) * np.sum(arr2))

    denominator = np.sqrt(
         (len(arr1) * np.sum(arr1 ** 2) - (np.sum(arr1)) ** 2) * 
         (len(arr2) * np.sum(arr2 ** 2) - (np.sum(arr2)) ** 2)
    )
    r = numerator / denominator
    
    if return_scatter:
        plt.scatter(arr1, arr2);
    return r.round(3)

## completed - hackerrank did not accept bc numpy 
```
